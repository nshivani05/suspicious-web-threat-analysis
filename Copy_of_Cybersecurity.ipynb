```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Load the dataset
df = pd.read_csv('CloudWatch_Traffic_Web_Attack.csv')

# Check basic info
print(df.shape)
df.head()
```

```python
# Data structure summary
df.info()

# Check for missing values
df.isnull().sum()
```

```python
# Convert string columns to datetime
df['creation_time'] = pd.to_datetime(df['creation_time'], errors='coerce')
df['end_time'] = pd.to_datetime(df['end_time'], errors='coerce')
df['time'] = pd.to_datetime(df['time'], errors='coerce')

# Confirm changes
df[['creation_time', 'end_time', 'time']].head()
```

```python
# Extract useful time features
df['date'] = df['time'].dt.date
df['hour'] = df['time'].dt.hour
df['day_of_week'] = df['time'].dt.day_name()
```

```python
# Group by hour
hourly_threats = df.groupby('hour').size().reset_index(name='count')

plt.figure(figsize=(10, 5))
sns.barplot(data=hourly_threats, x='hour', y='count', palette='coolwarm')
plt.title("Threats by Hour of the Day")
plt.xlabel("Hour")
plt.ylabel("Number of Interactions")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Group by date
daily_threats = df.groupby('date').size()

# Plot
plt.figure(figsize=(10, 4))
sns.lineplot(x=daily_threats.index, y=daily_threats.values, marker='o')
plt.title("Suspicious Web Interactions Per Day")
plt.xlabel("Date")
plt.ylabel("Number of Interactions")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Top 10 source IPs by count
top_ips = df['src_ip'].value_counts().head(10)

# Plot
plt.figure(figsize=(10, 5))
sns.barplot(y=top_ips.index, x=top_ips.values, palette='viridis')
plt.title("Top 10 Source IPs (Most Suspicious Interactions)")
plt.xlabel("Number of Events")
plt.ylabel("Source IP")
plt.show()
```

```python
# Top 10 countries
top_countries = df['src_ip_country_code'].value_counts().head(10)

plt.figure(figsize=(10, 5))
sns.barplot(y=top_countries.index, x=top_countries.values, palette='crest')
plt.title("Top Countries Generating Suspicious Traffic")
plt.xlabel("Number of Events")
plt.ylabel("Country Code")
plt.show()
```

```python
# Top 10 rule names
top_rules = df['rule_names'].value_counts().head(10)

plt.figure(figsize=(10, 5))
sns.barplot(y=top_rules.index, x=top_rules.values, palette='flare')
plt.title("Most Triggered Detection Rules")
plt.xlabel("Count")
plt.ylabel("Rule Name")
plt.show()
```

```python
# Top 10 most targeted destination ports
top_ports = df['dst_port'].value_counts().head(10)

plt.figure(figsize=(10, 5))
sns.barplot(y=top_ports.index.astype(str), x=top_ports.values, palette='cool')
plt.title("Top 10 Targeted Destination Ports")
plt.xlabel("Number of Events")
plt.ylabel("Destination Port")
plt.show()
```

```python
# Count of different protocols
protocol_counts = df['protocol'].value_counts()

plt.figure(figsize=(8, 4))
sns.barplot(y=protocol_counts.index, x=protocol_counts.values, palette='light:#5A9')
plt.title("Protocol Usage in Suspicious Traffic")
plt.xlabel("Count")
plt.ylabel("Protocol")
plt.show()
```

```python
# Threshold for anomaly detection
threshold = df['bytes_out'].quantile(0.95)

# Flag anomalies
df['is_anomaly'] = df['bytes_out'] > threshold

# Count
df['is_anomaly'].value_counts()
```

```python
# Plot anomalous traffic over time
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df, x='time', y='bytes_out', hue='is_anomaly', palette={True: 'red', False: 'blue'})
plt.title("Traffic Volume with Anomalies Highlighted")
plt.xlabel("Time")
plt.ylabel("Bytes Out")
plt.legend(title='Anomaly')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
